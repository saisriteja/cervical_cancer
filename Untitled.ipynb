```python
def load_obj(name):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
from tqdm import tqdm
```

```python
def getdata(f,files,imgs,shape,types):
    d = []
    for x in tqdm(f):
        #print(x)
        
        i = x.split()
        x = i[2:]
        x = [int(y) for y in x]
        name = i[0].split('\\')[1]
        position_file = files.index(name)
        #print(shapes[position_file])

        images,shape = imgs[position_file],shapes[position_file]
        w,h,_ = shape
        w,h = 200/w,200/h
        
        no = i[1]
        
        temp_data = []
        
        for j in range(int(no)):
            pp = [x[2*j]*w,x[2*j+1]*h,x[2*j+2]*w,x[2*j+3]*h]
            pp = [int(y) for y in pp]
            temp_data.append([name,pp,types])
            #print(name,pp)
        d.extend(temp_data)
        
    return d
```

```python
def custom_loss(y_true, y_pred):
  # target is a 8-tuple
  # (row, col, depth, width, class1, class2, class3, object_appeared)

  bce = binary_crossentropy(y_true[:, :4], y_pred[:, :4]) # location
  cce = categorical_crossentropy(y_true[:, 4:7], y_pred[:, 4:7]) # object class
  bce2 = binary_crossentropy(y_true[:, -1], y_pred[:, -1]) # object appeared

  return bce * y_true[:, -1] + cce * y_true[:, -1] + 0.5 * bce2
```

```python
def make_model():
  vgg = tf.keras.applications.VGG16(
    input_shape=[dims, dims, 3],
    include_top=False,
    weights='imagenet')
  x = Flatten()(vgg.output)
  x1 = Dense(4, activation='sigmoid')(x) # location
  x2 = Dense(3, activation='softmax')(x) # object class
  x3 = Dense(1, activation='sigmoid')(x) # object appeared
  x = Concatenate()([x1, x2, x3])
  model = Model(vgg.input, x)
#   model.compile(loss=custom_loss, optimizer=SGD(learning_rate=0.01, momentum=0.9))
  model.compile(loss=custom_loss, optimizer=Adam(lr=0.0001))
  return model
```

```python
dims = 200;
def image_generator(complete,batch_size=64):
    while True:
        for _ in range(50):
            X = np.zeros((batch_size, dims, dims, 3))
            Y = np.zeros((batch_size, 8))

            test1_data,files_imgs1,test2_data,files_imgs2,test3_data,files_imgs3 = complete  

            shuffle(test1_data)
            for i in range(0,batch_size//3):
                n = test1_data[i][0]
                #print()
                X[i] = np.array(files_imgs1[n])
                row0,col0,row1,col1 = test1_data[i][1]
                Y[i,0] = row0/dims
                Y[i,1] = col0/dims
                Y[i,2] = abs(row1 - row0)/dims
                Y[i,3] = abs(col1 - col0)/dims
                Y[i,4 + 0] = 1

            shuffle(test2_data)
            for i in range(batch_size//3,2*batch_size//3):
                n = test2_data[i][0]
                X[i] = np.array(files_imgs2[n])
                row0,col0,row1,col1 = test2_data[i][1]
                Y[i,0] = row0/dims
                Y[i,1] = col0/dims
                Y[i,2] = abs(row1 - row0)/dims
                Y[i,3] = abs(col1 - col0)/dims
                Y[i,4 + 1] = 1
                Y[i,7] = 1
            shuffle(test3_data)
            for i in range(2*batch_size//3,batch_size):
                n = test3_data[i][0]
                X[i] = np.array(files_imgs3[n])
                row0,col0,row1,col1 = test3_data[i][1]
                Y[i,0] = row0/dims
                Y[i,1] = col0/dims
                Y[i,2] = abs(row1 - row0)/dims
                Y[i,3] = abs(col1 - col0)/dims
                Y[i,4 + 2] = 1
                Y[i,7] = 1


            # did the pokemon appear?

        yield X / 255., Y
            
        
```

```python
comp = (test1_data,files_imgs1,test2_data,files_imgs2,test3_data,files_imgs3)
```
